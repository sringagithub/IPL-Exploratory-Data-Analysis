```python
#The number of cities
matches['city'].nunique()
```

```python
#The list of cities where the matches took place
matches['city'].unique()
```

```python
#Displayng the number of matches in each cities(countplot) 
data = matches.city.value_counts()
sns.barplot(y = data.index, x = data, orient='h');
```

```python
#From the above plot it is clear that the Mumbai city has hosted the maximum number of IPL matches
```

```python
#Displaying the teams
matches['team1'].unique()
```

```python
#The number of matches played by each team
matches['team1'].nunique()
```

```python
#The team which played most number of matches as team 1
data = matches.team1.value_counts()
sns.barplot(y = data.index, x = data, orient='h');
```

```python
#From the above barplot we can see that mumbai indians played the maximum number of matches as team 1
```

```python
#The team which played most number of matches as team 1
data = matches.team2.value_counts()
sns.barplot(y = data.index, x = data, orient='h');
```

```python
#From the above barplot we can see that Kolkata Knight Riders and Royal Challengers Bangalore played the maximum number of matches as team 2
```

```python
#The number of matches played in each seasons
sns.countplot(x='season', data=matches)
plt.show()
```

```python
#From the above countplot we can see that in the season 2013 the maximum number of matches took place followed by 2012 and then 2011.
```

```python
#From the below barplot it is clear that Mumbai Indians have won the maximum number of titles followed by Chennai Super Kings.
data = matches.winner.value_counts()
sns.barplot(y = data.index, x = data, orient='h');
```

```python
#Displaying an image of the Mumbai Indians team who won the maximum number of titles.
import matplotlib.image as mpimg
image = mpimg.imread("mi.png")
plt.title('Mumbai Indians')
plt.imshow(image)
plt.show()
```

```python
#Visualizing the top player of the match winners and we can see that Chris Gayle and AB de villiers are at top.
top_players = matches.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
sns.barplot(x = top_players.index, y = top_players, orient='v'); #palette="Blues");
plt.show()
```

```python
#Displaying an image of the both GAyle and De villiers.
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
image = mpimg.imread("gayle.jpg")
plt.title('Chris Gayle')
plt.imshow(image)
plt.show()
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
image = mpimg.imread("AB.jpg")
plt.title('AB De Villiers')
plt.imshow(image)
plt.show()
```

```python
#Dispalying the number of times when toss winner was the match winner
ss = matches['toss_winner'] == matches['winner']
ss.groupby(ss).size()
```

```python
#We can see that 393 times Match winner and toss winner was the same and hence we can say that there is a postive influence of winning the toss on winning the match.
```

```python
#Visualizing the how many times match winner and toss winner were the same using countplot.
sns.countplot(ss);
```
